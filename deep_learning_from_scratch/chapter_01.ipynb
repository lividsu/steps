# Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 6.283, 0.01)
y = np.cos(x)
plt.plot(x, y)
# plt.show()
```

```python
plt.plot(y, x)
```

```python
y.shape
```

```python
x[:10]
```

```python
x = np.arange(0, 20, 0.01)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='sin~')
plt.plot(x, y2, label='cos!', linestyle='--')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('this is title')
plt.legend()
plt.show()
```
